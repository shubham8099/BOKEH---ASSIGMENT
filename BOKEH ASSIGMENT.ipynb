Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, show 
from bokeh.io import output_notebook

# Prepare the Data
x = [1,2,3,4,5]
y = [6,7,2,4,8]

# Create the bokeh figure
plot = figure(title="Simple Scatter plot" , x_axis_label='X-axis' , y_axis_label='Y-axis')

# Add Glyphs (Scatter plot)
plot.circle(x,y,size=10 , color='navy' , alpha=0.5)

# Customixze the plot
plot.title.text_color = "olive"
plot.title.text_font = "times"
plot.title.text_font_style = "italic"

# Show the plot
output_notebook()
show(plot)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

* Let us Taking the Example for How we can customise the Bokeh :-

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Sample Data
x = [1,2,3,4,5]
y = [6,7,8,9,10]

# Crate the Bokeh figure
plot = figure(title="Customizing the Axis" , x_axis_label="X-axis" , y_axis_label="Y-axis")

# Add Circle Glyph to the Plot
plot.circle(x,y,size=10,color="navy" ,alpha=0.5)

# Customize The X-Axis
plot.xaxis.axis_label_text_color="orange"
plot.xaxis.major_label_text_color="orange"
plot.xaxis.axis_line_width = 2

# Customize the Y-Axis
plot.yaxis.axis_label_text_font_style = "italic"
plot.yaxis.major_label_orientation = "vertical"

# Customize the title
plot.title.text_color="olive"
plot.title.text_font="times"
plot.title.text_font_style="italic"

# Customize the legend
plot.legend.label_text_color="navy"
plot.legend.label_text_font="helvetica"
plot.legend.label_text_font_style="bold"

# Show the Plot
output_notebook()
show(plot)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

# Bokeh server is a componet of the bokeh library  that allow to us we can create the interactive web Applications with a Real-time update . it enables the creation of Dynamic and Responsive plot that can updated in response to user interactions or external events . By using Bokeh server we can Create the ineractive Dashboard, data exploration tools , and other applications where live update to the visualization are required .

* Let us taking the example of bokeh server :-


```python
from bokeh.plotting import figure, curdoc
from bokeh.io import output_notebook
from bokeh.layouts import column
from bokeh.models import Slider

# Create the Bokeh Figure and Data
plot = figure(title="Dynamic Scatter Plot" , width=500,height=400)
slider = Slider(start=1, end=10, value=1, step=1 , title="Number of Points")

def update_plot(attrname ,old,new):
    # Callback Function to update the plot based on slider value
    num_points = slider.value
    x = list(range(num_points))
    y = [i**2 for i in x]
    plot.circle(x,y,size=10)
    
# Attach the callback functions to the slider's value attributes
slider.on_change('value' , update_plot)

# Initial plot setup
update_plot('value' ,None, slider.value)

# Arrange the Plot and slider in a layout
layout = column(slider,plot)

# Set up the bokeh Document 
curdoc().add_root(layout)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Crate a bokeh plot
    plot = figure(title="Flask Bokeh Example" , width=500 , height=400)
    plot.circle([1,2,3,4,5] , [6,7,2,4,8] , size=10 , color ="navy" , alpha = 0.5)
    
    # Embed the plot into the HTML templates
    script , div = components(plot)
    return render_template('index.html' , script=script , div=div)
if __name__ == "__main__":
    app.run(debug=True)
```

```python
# Create the Flask templates 
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Flask bokeh Example</title>
    {{script | safe }}
</head>
<body>
    <h1>Bokeh Plot </h1>
    {{div | safe}}
</body> 
</html>

python app.py
```
